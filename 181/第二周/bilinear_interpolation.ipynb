```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread("lenna.png"), cv2.COLOR_BGR2RGB)
```

```python
plt.figure()
plt.imshow(img)
plt.show()
```

```python
high, wide, channels = img.shape
```

```python
new_high, new_wide = 800, 800
```

```python
new_img = np.zeros((new_wide, new_high, channels),img.dtype)
```

```python
scale_high, scale_wide = float(high) / new_high, float(wide) / new_wide
```

```python
for i in range(channels):
    for new_x in range(new_high):
        for new_y in range(new_wide):
            
            x = (new_x + 0.5) * scale_high - 0.5
            y = (new_y + 0.5) * scale_wide - 0.5
            
            x_0 = int(np.floor(x))
            x_1 = min(x_0 + 1, high - 1)
            y_0 = int(np.floor(y))
            y_1 = min(y_0 + 1, wide - 1)
            
            value_0 = (x_1 - x) * img[x_0, y_0, i] + (x - x_0) * img[x_1, y_0, i]
            value_1 = (x_1 - x) * img[x_0, y_1, i] + (x - x_0) * img[x_1, y_1, i]
            new_img[new_x, new_y,i] =int((y_1 - y) * value_0 + (y - y_0) * value_1)    
```

```python
plt.figure()
plt.imshow(new_img)
plt.show()
```
