```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lenna.png")
```

```python
high, wide = img.shape[:2]
```

```python
# 灰度化和二值化

gray = np.zeros((wide, high), img.dtype)  #  灰度图像模板
binary = np.zeros((wide, high))  # 二值图像模板
for x in range(high):
    for y in range(wide):
        pixels = img[x, y]
        gray[x, y] = int(pixels[0] * 0.11 + pixels[1] * 0.59 + pixels[2] * 0.3)
        if gray[x, y] >= 255/2:
            binary[x, y] = 1
```

```python
plt.figure()
plt.subplot(121)
plt.imshow(gray, cmap="gray")
plt.subplot(122)
plt.imshow(binary, cmap="binary")
plt.show()
```
