```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lenna.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
high, wide, channels = img.shape
```

```python
new_img = np.zeros((800, 800, channels), img.dtype)
```

```python
scale_high = 800 / high
scale_wide = 800 / wide
```

```python
for x in range(800):
    for y in range(800):
        i = int(x / scale_high)
        j = int(y / scale_high)
        new_img[x, y] = img[i, j]
```

```python
plt.figure()
plt.imshow(img)
plt.show()
```

```python
plt.figure()
plt.imshow(new_img)
plt.show()
```
